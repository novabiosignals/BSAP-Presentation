# Types of signal and their intrinsic properties

```python
fig, ax = subplots(2,2)

#continuous
t_sin = np.linspace(0, 8, 500)
y_sin = np.sin(2*t_sin)
ax[0,0].plot(t_sin, y_sin, linewidth=3, color='k')
ax[0,0].set_ylabel('amplitude',color='grey')
ax[0,0].set_xlabel('time(s)',color='grey')
axis([0, 8, -2, 2])
ax[0,0].set_title('Continuous', fontsize=12, color='grey')

#periodical
[t,s]= loadtxt('data/cleanecg.txt.gz')
ax[0,1].set_xlabel('time (s)',color='grey')
ax[0,1].set_ylabel('amplitude (mV)',color='grey')
ax[0,1].plot(t,s, color='k')
axis([0,6,-300,300])
ax[0,1].set_title('Periodical', fontsize=12, color='grey')

#discrete
y_discrete = np.round(7*y_sin)
ax[1,0].plot(t_sin, y_discrete, linewidth=3, color='k')
ax[1,0].set_ylabel('amplitude',color='grey')
ax[1,0].set_xlabel('time(s)',color='grey')
axis([0, 8, -8, 8])
ax[1,0].set_title(u'Discrete', fontsize=12, color='grey')

#non-periodical
[t,s]= loadtxt('data/emg.txt.gz')
ax[1,1].set_xlabel('time (s)',color='grey')
ax[1,1].set_ylabel('amplitude (mV)',color='grey')
axis([0, 4.5, -3, 3])
ax[1,1].set_title(u'Non-Periodical', fontsize=12, color='grey')

ax[1,1].plot(t[::2],s[::2], color='k')

tight_layout()
```

### Characteristics in time domain

```python
from matplotlib.ticker import NullFormatter

nullfmt=NullFormatter()

[t,s]= loadtxt('data/emg.txt.gz')

figure()

ax1=axes([0.125,0.1,0.5,0.8])
ax2=axes([0.125+0.5,0.1,0.2,0.8])


ax1.plot(t,s, color='k')
ax1.set_xticks(ax1.get_xticks()[:-1])
ax1.axis([0,5,ax1.axis()[2],ax1.axis()[3]])


meanv=s.mean()
mi=s.min()
mx=s.max()
sd=s.std()


ax2.bar(-0.1,mx-mi,1,mi,lw=2,color='#f0f0f0', alpha=0.5, edgecolor='k')
ax2.bar(-0.1,sd*2,1,meanv-sd,lw=2,color='#a0a0a0', alpha=0.5, edgecolor='k')
ax2.bar(-0.1,0.2,1,meanv-0.1,lw=2,color='#202020', alpha=0.5, edgecolor='k')
ax2.axis([-1,1,ax1.axis()[2],ax1.axis()[3]])

ax2.yaxis.set_major_formatter(nullfmt)
ax2.set_xticks([])
```

### Fourier Series

```python
import matplotlib.gridspec as gridspec

t = arange(0.0, 8, 0.05)
x1 = sin(t*2) 

nullfmt=NullFormatter()

seed(230)

fig = figure(figsize=(20,10))

outerGrid = gridspec.GridSpec(1, 2, wspace=0.2, hspace=0.2)

#Plot sine waves decomposition

inner1 = gridspec.GridSpecFromSubplotSpec(10, 1, subplot_spec=outerGrid[0], wspace=0.1, hspace=0.1)

vx=[]

for i in range(9):
    v=sin(t*rand(1)*80+rand(1)*150)*rand(1)
    vx+=[v]
    ax=Subplot(fig, inner1[i])
    ax.yaxis.set_major_formatter(nullfmt)
    ax.xaxis.set_major_formatter(nullfmt)

    ax.plot(t,v,color='grey')
    ax.axis((0,8,-3,3))
    fig.add_subplot(ax)

ax=Subplot(fig, inner1[9])
ax.yaxis.set_major_formatter(nullfmt)  
vx=array(vx)
ax.plot(t,vx.sum(0), 'k')
ax.axis((0,8,-3,3))
fig.add_subplot(ax)

#Plot frequency information
fmax = 1000
fftax = Subplot(fig, outerGrid[1])

fs=abs(sp.fft(vx.sum(0)))
f=linspace(0,fmax/2,len(vx.sum(0))/2)
fftax.plot(f[1:int(len(vx.sum(0))/2)],fs[1:int(len(vx.sum(0))/2)], linewidth=3, color='k')
fftax.set_ylabel('amplitude',color='grey')
fftax.set_xlabel('frequency (Hz)',color='grey')
fig.add_subplot(fftax)
```

### Square Wave

```python
def plotfft(s, fmax):
    fs=abs(fft(s))
    f=linspace(0,fmax/2,len(s)/2)
    plot(f[1:int(len(s)/2)],fs[1:int(len(s)/2)], linewidth=2, color='k')

t = arange(0.0, 8, 0.05)
x1 = sin(t*2) 

xx=sign(x1)

subplot(4,2,1)
plot(t,xx, linewidth=2, color='k')
f=fft(xx)
axis((0,8,-1.2,1.2))
subplot(4,2,2)
plotfft(xx,1000)

subplot(4,2,3)
f[50:750]=0
plot(t,ifft(f), linewidth=2, color='k')
axis((0,8,-1.2,1.2))
subplot(4,2,4)
plotfft(sp.ifft(f),1000)


subplot(4,2,5)
f[20:780]=0
plot(t,ifft(f), linewidth=2, color='k')
axis((0,8,-1.2,1.2))
subplot(4,2,6)
plotfft(ifft(f),1000)


subplot(4,2,7)
f[5:795]=0
plot(t,ifft(f), linewidth=2, color='k')
axis((0,8,-1.2,1.2))
subplot(4,2,8)
plotfft(ifft(f),1000)

tight_layout()
```

# Other Sensors

### Light

```python
s = loadtxt("data/lux.txt")[:,2]
n = 16
fs = 100
s = (s/(2**n))*100
t = linspace(0, len(s) / fs, len(s))
s = smooth(s, 10)

figure()
xlabel('time (s)', color='grey')
ylabel('amplitude', color='grey')
title('Lux', fontsize=12, color='grey')

plot(t, s, 'k', linewidth=2)
```

### Temperature

```python
#temperature
s = loadtxt("data/temp.txt")[:,2]
s = smooth(s, 100)
fs = 1000

#temperature conversion from analog to Celsius
#can be found on the following datasheet: http://www.biosignalsplux.com/datasheets/TMP_Sensor_Datasheet.pdf
vcc = 3
n = 16
ntcV = (s*vcc)/(2**n)
ntcR = (10000 * ntcV) / (vcc-ntcV)
a0 = 0.00112764514
a1 =0.000234282709
a2 = 0.0000000877303013
tempK = 1/(a0+a1*log(ntcR)+a2*(log(ntcR))**3)
tempT = tempK - 273.15

t = linspace(0, len(s)/fs, len(s))

figure()
xlabel('time (s)', color='grey')
ylabel('amplitude', color='grey')
title('Temperature', fontsize=12, color='grey')

plot(t, tempT, linewidth=2, color='k')
```

### Goniometer

```python
s = loadtxt("data/goniometer.txt")[:, 3]

#conversion from analog to angle degrees: http://www.biosignalsplux.com/datasheets/GON_Sensor_Datasheet.pdf
n = 16
fs = 100
vcc =3
s = (((vcc/ ((2 ** n)-1)) * s)-vcc/2)/((vcc/2)*606*0.00001)
s = smooth(s, 10)
t = linspace(0, len(s) / fs, len(s))


figure()
xlabel('time (s)', color='grey')
ylabel('amplitude', color='grey')
title('Goniometer', fontsize=12, color='grey')

plot(t, s, linewidth=2, color='k')
```
