## Sampling Theorem

```python
ff = [10, 3.1, 1.003, 0.556]  # the frequency of the signal
fs1 = 100  # sample rate
fs = 10
time = 5

i = 0

for f in ff:
    i += 1

    a = subplot(4, 1, i)

    t = np.linspace(0, time, int(fs * time), endpoint=False)
    t1 = np.linspace(0, time, int(fs1 * time), endpoint=False)
    y = np.cos(2 * np.pi * (fs/f) * t)
    y1 = np.cos(2 * np.pi * (fs/f) * t1)

    a.plot(t1, y1, linewidth=2, color='grey')
    a.plot(t, y, 'o', color='k', markersize=5)
    if(i==1):
        a.set_ylabel(r'$F_{sp} = $'+str(int(f)) + r"$ *F_{signal}$", fontsize=12)
    elif(i==4):
        a.set_ylabel(r'$F_{sp} \approx F_{signal}/2$', fontsize=12)
    else:
        a.set_ylabel(r'$F_{sp} \approx $'+str(int(f))+r"$ *F_{signal}$", fontsize=12)
tight_layout()
```

## Quantization 

```python
t_sin = np.linspace(0, 8, 500)
y_sin = np.sin(2*t_sin)
y_discrete = np.round(7*y_sin)
plot(t_sin, y_discrete, linewidth=3, color='k')
ylabel('amplitude',color='grey')
xlabel('time(s)',color='grey')
axis([0, 8, -8, 8])
title(u'Quantization', fontsize=12, color='grey')
```
